# Atmpspheric pressure data preparation

```r
# load required packages
library(dplyr)
library(lubridate)

# Read the data
data <- read.csv("../data/FLX_CH-Oe2_FLUXNET2015_FULLSET_2004-2023_1-3/FLX_CH-Oe2_FLUXNET2015_FULLSET_DD_2004-2023_1-3.csv")
colnames(data)

# Convert the date column to Date type
data$TIMESTAMP <- as.Date(as.character(data$TIMESTAMP), format="%Y%m%d")

# Create new dataframe with date and atmospheric pressure
pa_df <- data %>%
    select(TIMESTAMP, PA_F) %>%
    #modify the name
    rename(PA_F_kpa = PA_F) %>%
    # add a new column for atmospheric pressure in pascals
    mutate(PA_F_pa = PA_F_kpa * 1000)

# Show first few rows
head(pa_df)

# Save to CSV
write.csv(pa_df, "../data/01_data_prep/05_pa.csv", row.names = FALSE)
```

```r
# Load ggplot2 and other required packages for plotting
library(ggplot2)
library(tidyr)
library(gridExtra)

# Add year, month, and season columns for plotting
pa_df$year <- year(pa_df$TIMESTAMP)
pa_df$month <- month(pa_df$TIMESTAMP)
pa_df$season <- case_when(
    pa_df$month %in% c(12, 1, 2) ~ "Winter",
    pa_df$month %in% c(3, 4, 5) ~ "Spring",
    pa_df$month %in% c(6, 7, 8) ~ "Summer",
    pa_df$month %in% c(9, 10, 11) ~ "Fall"
)

# Reshape data to long format for plotting both units
data_long <- pa_df %>%
    select(TIMESTAMP, year, PA_F_kpa, PA_F_pa) %>%
    pivot_longer(cols = c(PA_F_kpa, PA_F_pa), 
                 names_to = "pa_type", 
                 values_to = "pa_value")

# Create main time series plot for atmospheric pressure in kPa
plot_main <- ggplot(pa_df, aes(x = TIMESTAMP, y = PA_F_kpa)) +
    geom_line(color = "darkblue", alpha = 0.7, linewidth = 0.5) +
    scale_x_date(date_labels = "%Y", date_breaks = "1 years") +
    labs(title = "Atmospheric Pressure Time Series Over Time",
         x = "Year",
         y = "Atmospheric Pressure (kPa)",
         subtitle = "Daily atmospheric pressure measurements") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1),
          plot.title = element_text(size = 14, face = "bold"))

# Create seasonal comparison plot
seasonal_data <- pa_df %>%
    filter(!is.na(PA_F_kpa)) %>%
    group_by(season, year) %>%
    summarise(mean_pa = mean(PA_F_kpa, na.rm = TRUE), .groups = "drop") %>%
    mutate(date = as.Date(paste(year, "01", "01", sep = "-")))

plot_seasonal <- ggplot(seasonal_data, aes(x = date, y = mean_pa, color = season)) +
    geom_line(alpha = 0.8, linewidth = 0.7) +
    scale_color_manual(values = c("Winter" = "blue", "Spring" = "green", 
                                  "Summer" = "red", "Fall" = "orange")) +
    scale_x_date(date_labels = "%Y", date_breaks = "1 years") +
    labs(title = "Seasonal Atmospheric Pressure Patterns",
         x = "Year",
         y = "Mean Atmospheric Pressure (kPa)",
         color = "Season") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1),
          legend.position = "bottom",
          plot.title = element_text(size = 14, face = "bold"))

# Create annual summary plot
annual_summary <- pa_df %>%
    filter(!is.na(PA_F_kpa)) %>%
    group_by(year) %>%
    summarise(
        mean_pa = mean(PA_F_kpa, na.rm = TRUE),
        max_pa = max(PA_F_kpa, na.rm = TRUE),
        min_pa = min(PA_F_kpa, na.rm = TRUE),
        .groups = "drop"
    ) %>%
    pivot_longer(cols = c(mean_pa, max_pa, min_pa),
                 names_to = "stat_type",
                 values_to = "pa_value")

plot_annual <- ggplot(annual_summary, aes(x = year, y = pa_value, color = stat_type)) +
    geom_line(alpha = 0.8, linewidth = 0.7) +
    geom_point(alpha = 0.6, size = 1.5) +
    scale_color_manual(values = c("mean_pa" = "darkblue", 
                                  "max_pa" = "red", 
                                  "min_pa" = "blue"),
                       labels = c("Annual Mean", "Annual Maximum", "Annual Minimum")) +
    labs(title = "Annual Atmospheric Pressure Statistics",
         x = "Year",
         y = "Atmospheric Pressure (kPa)",
         color = "Statistic") +
    theme_minimal() +
    theme(legend.position = "bottom",
          plot.title = element_text(size = 14, face = "bold"))

# Create comparison plot for kPa vs Pa units
plot_comparison <- ggplot(data_long, aes(x = TIMESTAMP, y = pa_value, color = pa_type)) +
    geom_line(alpha = 0.7, linewidth = 0.5) +
    scale_color_manual(values = c("PA_F_kpa" = "darkblue", "PA_F_pa" = "darkgreen"),
                       labels = c("Pressure (kPa)", "Pressure (Pa)")) +
    scale_x_date(date_labels = "%Y", date_breaks = "1 years") +
    facet_wrap(~pa_type, scales = "free_y", ncol = 1,
               labeller = labeller(pa_type = c("PA_F_kpa" = "kPa", "PA_F_pa" = "Pa"))) +
    labs(title = "Atmospheric Pressure in Different Units",
         x = "Year",
         y = "Atmospheric Pressure",
         color = "Unit") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1),
          legend.position = "bottom",
          plot.title = element_text(size = 14, face = "bold"))

# Display individual plots
print(plot_main)
print(plot_seasonal)
print(plot_annual)
print(plot_comparison)

# Create the directory for trend plots if it does not exist
if (!dir.exists("../data/01_data_prep/trend_plots")) {
    dir.create("../data/01_data_prep/trend_plots", recursive = TRUE)
}

# Save individual plots
ggsave("../data/01_data_prep/trend_plots/05_pa_main_series.png", 
       plot = plot_main, width = 12, height = 6, bg = "white", dpi = 300)

ggsave("../data/01_data_prep/trend_plots/05_pa_seasonal.png", 
       plot = plot_seasonal, width = 12, height = 6, bg = "white", dpi = 300)

ggsave("../data/01_data_prep/trend_plots/05_pa_annual.png", 
       plot = plot_annual, width = 10, height = 6, bg = "white", dpi = 300)

ggsave("../data/01_data_prep/trend_plots/05_pa_comparison.png", 
       plot = plot_comparison, width = 12, height = 8, bg = "white", dpi = 300)

# Create combined plot
plot_combined <- grid.arrange(plot_main, plot_seasonal, plot_annual, plot_comparison, ncol = 1)

# Save the combined plot
ggsave("../data/01_data_prep/trend_plots/05_pa_combined.png", 
       plot = plot_combined, width = 12, height = 20, bg = "white", dpi = 300)
```
