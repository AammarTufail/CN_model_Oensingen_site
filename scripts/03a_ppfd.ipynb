```r
##########################################################
# ---- Read HH data and take averages based on day and night times ----
##########################################################

# 01_vpd_daytime.R
# This script reads half-hourly data, extracts daytime average vapour pressure deficit (VPD_F_DAY),
# and writes the result to a CSV file.

# ---- Package Installation and Loading ----
# Ensure required packages are installed and loaded
required_packages <- c("dplyr", "lubridate")
installed_packages <- rownames(installed.packages())
for (pkg in required_packages) {
    if (!pkg %in% installed_packages) {
        install.packages(pkg, dependencies = TRUE)
    }
    library(pkg, character.only = TRUE)
}

# ---- Data Reading ----
# Read the half-hourly measurement file (CSV)
data <- read.csv("../data/FLX_CH-Oe2_FLUXNET2015_FULLSET_2004-2023_1-3/FLX_CH-Oe2_FLUXNET2015_FULLSET_DD_2004-2023_1-3.csv")
head(data)      # Preview the first few rows
colnames(data)  # Display column names

# ---- Extract PPFD columns and save to CSV ----
# Select date and PPFD-related columns
ppfd_data <- data %>%
    select(TIMESTAMP, PPFD_IN, PPFD_IN_QC, PPFD_DIF, PPFD_DIF_QC)

# Convert TIMESTAMP to Date format and replace it
ppfd_data$date <- as.Date(as.character(ppfd_data$TIMESTAMP), format = "%Y%m%d")
ppfd_data <- ppfd_data %>%
    select(-TIMESTAMP) %>%
    select(date, everything())

# Write to CSV file
write.csv(ppfd_data, "../data/01_data_prep/03_ppfd_data.csv", row.names = FALSE)

# Preview the extracted data
head(ppfd_data)


# Load ggplot2 for plotting
library(ggplot2)

# Add year column for plotting
ppfd_data$year <- year(ppfd_data$date)

# Reshape data to long format for plotting multiple variables
library(tidyr)
data_long <- ppfd_data %>%
    select(date, year, PPFD_IN, PPFD_DIF) %>%
    pivot_longer(cols = c(PPFD_IN, PPFD_DIF), 
                 names_to = "ppfd_type", 
                 values_to = "ppfd_value")

# Create a single plot with different colors for each PPFD type
plot <- ggplot(data_long, aes(x = date, y = ppfd_value, color = ppfd_type)) +
            geom_line(alpha = 0.7, size = 0.5) +
            scale_color_manual(values = c("PPFD_IN" = "red", 
                                        "PPFD_DIF" = "blue"),
                            labels = c("PPFD Incoming", "PPFD Diffuse")) +
            scale_x_date(date_labels = "%Y", date_breaks = "1 years") +
            labs(title = "Comparison of Different PPFD Variables Over Time",
                x = "Year",
                y = "PPFD (μmol m⁻² s⁻¹)",
                color = "PPFD Type") +
            theme_minimal() +
            theme(axis.text.x = element_text(angle = 45, hjust = 1),
                legend.position = "bottom");plot

# Create individual plots for each PPFD variable
plot_in <- ggplot(ppfd_data, aes(x = date, y = PPFD_IN)) +
    geom_line(color = "red", alpha = 0.7, size = 0.5) +
    scale_x_date(date_labels = "%Y", date_breaks = "1 years") +
    labs(title = "PPFD Incoming Over Time",
            x = "Year",
            y = "PPFD IN (μmol m⁻² s⁻¹)") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))

plot_dif <- ggplot(ppfd_data, aes(x = date, y = PPFD_DIF)) +
    geom_line(color = "blue", alpha = 0.7, size = 0.5) +
    scale_x_date(date_labels = "%Y", date_breaks = "1 years") +
    labs(title = "PPFD Diffuse Over Time",
            x = "Year",
            y = "PPFD DIF (μmol m⁻² s⁻¹)") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))

# Display the plots
print(plot_in)
print(plot_dif)

# Stack all plots together for a comprehensive view
library(gridExtra)
plot_combined <- gridExtra::grid.arrange(plot, plot_in, plot_dif, ncol = 1)
plot_combined

# Create the directory for trend plots if it does not exist
if (!dir.exists("../data/01_data_prep/trend_plots")) {
    dir.create("../data/01_data_prep/trend_plots", recursive = TRUE)
}

# Save the combined plot as a PNG file
ggsave("../data/01_data_prep/trend_plots/03a_ppfd_comparison_plots.png", plot = plot_combined, width = 20, height = 15, bg = "white", dpi = 300)

# stack all single plots together
plots_stacked <- gridExtra::arrangeGrob(plot, plot_in, plot_dif, ncol = 1)
plots_stacked
# Save the combined plot as a PNG file
ggsave("../data/01_data_prep/trend_plots/03b_ppfd_combined_plots.png", plot = plots_stacked, width = 20, height = 15, bg = "white", dpi = 300)
```
